```python
from sklearn import svm
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import Pipeline
from sklearn import cross_validation
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import KFold
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from sklearn.pipeline import Pipeline

import datetime
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# loading data
train_path = "train.csv"
test_path = "test.csv"
raw_data = pd.read_csv(train_path, header=0)
raw_test_data = pd.read_csv(test_path, header=0)
raw_data['niecount'] = raw_data['count']
```

```python
#feature engineering 
# definiuje funkcje zeby latwo powtozyc to samo dla zbioru testowego
def enhance_data(raw_data, shuffle=True):
    # permutuje sobie obserwacje bo przy cross-validation bede bral 
    # pierwszych 80% do zbioru treningowego, koljenych 20% do testowego
    # a te dane przyszly posortowane chronologicznie wiec do zbioru 
    # testowego trafilyby tylko dane z ostatnich miesiecy itp.
    if shuffle:
        data = raw_data.reindex(np.random.permutation(raw_data.index))
    else:
        data = raw_data.copy()
    # zamieniam numeryczna ceche "season" na 4 cechy boolowskie 
    seas_dict = {1: 'winter', 2: 'spring', 3: 'summer', 4: 'fall'}
    for elem in data.season.unique():
        data[seas_dict[elem]] = data['season'] == elem
        
    # analogicznie z "weather" - oznaczenia przeczytane z kaggla
    weather_dict = {1:'clear', 2: 'mist', 3: 'snow', 4: 'rain'}
    for elem in data.weather.unique():
        data[weather_dict[elem]] = data['weather'] == elem
    
    # zamieniam date ktora jest stringiem na prawdziwy obiekt typu data
    data['datetime_real'] = [datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") for x in data.datetime]
    data['hour'] = [x.hour for x in data.datetime_real]
    data['weekday'] = [x.isoweekday() for x in data.datetime_real]
    start = datetime.datetime(2012, 1, 1)
    # dodaje ceche - znormalizowanyc czas 
    data['norm_time'] = [(x-start).days/360. for x in data.datetime_real]
    return data

data = enhance_data(raw_data)
test_data = enhance_data(raw_test_data, shuffle=False)
```

```python
# jak wyglada srednia wartosc "niecount" dla roznych sezonow, temperatur itd
# nie mam jeszcze dobrze rozkminionego plotowania w pythonie, ale to jest na pewno
# duzo lepsze niz nic
# w ipythonie nie zapomnij cell -> current ouput -> toggle scrolling
keys = ["season", "holiday", "workingday", "weather", "datetime_real", "temp", "atemp", 
            "humidity", "windspeed", "norm_time", "hour", "weekday"]

for key in keys:
    means = data.groupby(key).niecount.mean()
    counts = data.groupby(key).niecount.count()
    marker_sizes = [max(x, 15) for x in counts.get_values()]
    plt.figure(figsize=(20,10))
    fig = plt.scatter(means.keys(), means.get_values(), s=marker_sizes)
    # robie show() po kazdym plocie, bo inaczej wszystkie by sie narysowaly w jednym miejscu
    # alternatywnie mozna by uzyc sublotow, ale tak jest latwiej
    fig.axes.set_xlabel(key, fontsize=18)
    plt.show()

plt.show()
```

```python
n = len(data)
# wybierzmy ktorych featurow chcemy uzyc
# mozna DUZO lepsze! sama sproboj
feats = ['winter', 'spring', 'summer', 'temp', 'atemp', 'windspeed', 'humidity']
X = np.array(data.get(feats))
y = np.array(data.niecount)

test_n = len(test_data)
test_X = np.array(test_data.get(feats))
```

```python
# bardzo wazny krok - funkcja ktora oceni model
def score(clf):
    # k-fold czyli wez 4/5 do treningowego, reszte do testowego
    # i tak 5 razy za kazdym birac inna 1/5
    kf = KFold(n, n_folds=5)
    # powstanie 5 wynikow
    scores = []
    for train, test in kf:
        # X[train], X[test] bo train i test sa arrayami boolowskimi
        # czyli [True, False, True ...]
        # jak zrobisz np.array([1,2,3,4,5])[np.array([True, False, True, False, True])]
        # to dostaniesz array([1, 3, 5])
        clf.fit(X[train], y[train])
        # wiekszosc modeli regresji, jesli nie wszystkie, maja wbudowana funkcje score, 
        # jakbysmy chcieli to moglibysnt uzyc innej ze sklearn
        scores.append(clf.score(X[test], y[test]))
    return sum(scores)/len(scores)
```

```python
# a to najprzyjemniejsza czesc - sprawdzmy, jak rozne modele sobie radza
print score(LinearRegression())
print score(DecisionTreeRegressor(max_depth=8))
print score(AdaBoostRegressor(DecisionTreeRegressor(max_depth=3), n_estimators=30, random_state=0))
print score(AdaBoostRegressor(LinearRegression()))
print score(AdaBoostRegressor(DecisionTreeRegressor()))
```

```python
# bierzemy najlepszy z modeli i fitujemy go na serio
clf = LinearRegression()
clf.fit(X, y)
# a potem uzywamy na prawdziwym kagglowym zbiorze testowym
test_y = clf.predict(test_X)
```

```python
# i zapisujemy wyniki do pliku w takim formacie, o jaki prosili
result_df = pd.DataFrame({'datetime': test_data.datetime, "count": [int(round(y)) for y in  test_y]})
result_df.to_csv(path_or_buf="submit.csv", columns=['datetime', 'count'], index=False)
```
